```python
df_v1 = df[['title','overview','attributes','attribute_scope_notes', 'attribute_groups', 'attribute_group_scope_notes','l2s',
   'l3_scope_notes','l2_scope_notes','l1s', 'l1_scope_notes','l3s']]

df_v1 = df_v1.fillna('0')

from gensim.parsing.preprocessing import remove_stopwords

def clean_string(x):
    x = x.replace('\n','').replace('[','').replace(']','').replace("'",'').replace('&','').replace('0','').strip().replace('"','').replace('.','').replace("'",'')
    return remove_stopwords(x)

for col in df_v1.columns:
    print(col)
    df_v1[col] = df_v1[col].apply(lambda x: clean_string(x))
    
df_v1['tokens'] = df_v1.apply(' '.join, axis=1)
df_v1['tokens'] = df_v1['tokens'].apply(lambda x: clean_string(x))

df_v1['target'] = df['l3_ontology_ids']
df_v1['target'] = df_v1['target'].apply(lambda x: clean_string(x))

df_v2 = df_v1[['tokens','target']]
```
